*Modules Importing*

```python
import pandas as pd
```

```python
from sklearn.datasets import load_diabetes
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=load_diabetes()
```

```python
df
```

```python
dataset=pd.DataFrame(df.data)
```

```python
dataset.columns=df.feature_names
```

```python
dataset.head()
```

*Dependent&Independent Features*

```python
x=dataset
y=df.target
```

```python
y
```

*Train_Test_Split*

```python
from sklearn.model_selection import train_test_split
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
x_train
```

*Standardizing Dataset*

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.inverse_transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
regression=LinearRegression()
```

```python
from sklearn.model_selection import cross_val_score
regression = LinearRegression()
cross_val_score(regression,x_train,y_train,scoring='neg_mean_squared_error',cv=10)
regression
regression.fit(x_train,y_train)
```

```python
# Option 1: Define the variable `mse` in the current cell
mse = np.mean(cross_val_score(regression,x_train,y_train,scoring='neg_mean_squared_error',cv=10))

# Option 2: Make sure that the variable `mse` is defined in a way that makes it accessible to the current cell
# For example, you could define the variable `mse` in a previous cell and then use the `global` keyword to make it accessible to the current cell.
global mse
mse = np.mean(cross_val_score(regression,x_train,y_train,scoring='neg_mean_squared_error',cv=10))
```

```python
mse
```

```python
reg_pred=regression.predict(x_test)
```

```python
reg_pred
```

```python
import seaborn as sns
sns.displot(reg_pred-y_test,kind='kde')
```

```python
from sklearn.metrics import r2_score
score=r2_score(reg_pred,y_test)
score
```
